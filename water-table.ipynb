### Quais são as minas mais velhas do data frame? Qual é a qualidade da água delas?

```python
### Há várias minas com ano de construção igual a zero

nonInformedConstructionYear = data["construction_year"] == 0;
data[nonInformedConstructionYear]
```

```python
# Obtendo ano mais antigo registrado:
informedConstructionYear = data["construction_year"] > 0;
oldestYear = data[informedConstructionYear].sort_values(by="construction_year").construction_year.min();
oldestYear
```

```python
# Obtendo a quantidade de minas construídas no ano mais antigo:
builtInOldestYear = data.construction_year == oldestYear
data[builtInOldestYear].construction_year.count()
```

```python
# Obtendo a qualidade de água delas:

data[builtInOldestYear].groupby("water_quality")[["water_quality"]].describe()
```
